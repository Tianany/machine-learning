```python
df= pd.read_csv("train.csv")
df.head()
```

```python
x = df.iloc[:, 0:20]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
backward_feature_selector=SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),k_features=10,forward=False,floating=False,verbose=2,scoring='accuracy',cv=5).fit(x_train,y_train)
```

```python
backward_feature_selector.k_feature_idx_
```

```python
backward_feature_selector.k_feature_names_
```

```python
backward_feature_selector.k_score_
```
