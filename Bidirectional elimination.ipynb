```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE, RFECV
```

```python
# Load the data
df = pd.read_csv('traiin.csv')
```

```python
#from sklearn.preprocessing import LabelEncoder
#le= LabelEncoder()
#df['ID_code']=le.fit_transform(df['ID_code'])
#df
```

```python
# Split the data into features (X) and target (y)
X = df.drop('price_range', axis=1)
y = df['price_range']
```

```python
# Recursive Feature Elimination (RFE)
model = LinearRegression()
rfe = RFE(model, n_features_to_select=1)
rfe = rfe.fit(X, y)
```

```python
# Print the ranking of the features
print('Feature Ranking:')
for i, rank in enumerate(rfe.ranking_):
    print(f"{df.columns[i]}: Rank {rank}")
```

```python
# Recursive Feature Elimination with Cross-Validation (RFECV)
rfecv = RFECV(estimator=LinearRegression(), step=1, cv=5, scoring='r2')
rfecv.fit(X, y)
```

```python
# Print the number of selected features
print(f'\nOptimal number of features: {rfecv.n_features_}')
```

```python
# Print the selected feature names
print('Selected Features:')
for i, is_selected in enumerate(rfecv.support_):
    if is_selected:
        print(df.columns[i])
```
