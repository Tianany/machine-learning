```python
#import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

**LOAD THE DATASET**

```python
# Load the dataset
df_authors = pd.read_csv("metabolic_syndrome_synthetic_data.csv")
df_authors
```

**DATA STANDARDIZATION**

**CONVERT FROM STRING TO FLOAT**

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df_authors['sex']=le.fit_transform(df_authors['sex'])
```

```python
# Standardize the data
scaler = StandardScaler()
DNP_authors_standardized = scaler.fit_transform(df_authors)
df_authors_standardized = pd.DataFrame(DNP_authors_standardized, columns=['WC','TG','HDLC','SBP','DBP','FPG','sex'])
```

**CREATING A K-MODEL**

```python
# Create a K-Means model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
```

**FIT MODEL TO STANDARDIZED DATA**

```python
#droping rows which contain NAN values
#make use of the dropna function from pandas
df_authors_standardized.dropna(inplace=True)
```

```python
# Fit the model to the standardized data
kmeans.fit(df_authors_standardized)
```

**GET CLUSTER LABELS**

```python
# Get the cluster labels
labels = kmeans.labels_
```

**ADD CLUSTER LABELS TO ORIGINAL DATAFRAME**

```python
df_authors.dropna(inplace=True)
```

```python
# Add the cluster labels to the original dataframe
df_authors["cluster"] = labels
```

**PRINT CLUSTER LABELS**

```python
# Print the cluster labels
print(df_authors["cluster"].unique())
```
