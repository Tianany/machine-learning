```python
x=df.iloc[:,0:20]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
#splitting the model into x and y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=90)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                  k_features= 6,
                                                  forward=True,
                                                  floating=False,
                                                  verbose=2,
                                                  scoring="accuracy",
                                                  cv= 5).fit(x_train, y_train)
```

```python
# identifying which features were chosen as the most relevant for the model.
```

```python
forward_feature_selection.k_feature_idx_
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```

```python
#showing the data of featured predictions,at the end transposes the DataFrame, swapping the rows and columns.#
```

```python
pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```

```python
# Sequential Feature Selector  select the most important features from the input data (x_train) using a Random Forest classifier.
#It will select 6 features using the forward selection method, and the selection process will be evaluated using 
# 5-fold cross-validation with the accuracy scoring metric. 
#The verbosity level is set to 2, providing detailed output during the feature selection process
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                  k_features= (1,20),
                                                  forward=True,
                                                  floating=False,
                                                  verbose=2,
                                                  scoring="accuracy",
                                                  cv= 5).fit(x_train, y_train)
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```
