# DATA PRE-PROCESSING

## import the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
df= pd.read_csv("TRAIN.csv")
df
```

## clean the data

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cmap="viridis")
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['ID_code']=le.fit_transform(df['ID_code'])
```

```python
df
```

```python
y=df['target']
```

```python
y
```

```python
x=df.drop('target', axis=1)
```

```python
x
```

## split into training and test sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# MODELLING

## Buid and Train the model

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
```

## Make Predictions

```python
y_lr_train_pred= lr.predict(x_train)
y_lr_test_pred= lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

# EVALUATION

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse= mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2= r2_score(y_train, y_lr_train_pred)

lr_test_mse= mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2= r2_score(y_test, y_lr_test_pred)
```

```python
lr_results=pd.DataFrame(['Linear Regressiom',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns=('method', 'train mse', 'train r2', 'test mse', 'test r2')
```

```python
lr_results
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_pred= rf.predict(x_train)
y_rf_test_pred= rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse= mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2= r2_score(y_train, y_rf_train_pred)

rf_test_mse= mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2= r2_score(y_test, y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns=('method', 'train mse', 'train r2', 'test mse', 'test r2')
```

```python
rf_results
```
